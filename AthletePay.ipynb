```python
def calculate_inflation(base_year_data, current_year_data):
    # Calculate the CPI for the base year
    cpi_base_year = 100
    # Calculate the CPI for the current year
    cpi_current_year = (sum(current_year_data) / sum(base_year_data)) * 100
    # Calculate the inflation rate
    inflation_rate = ((cpi_current_year - cpi_base_year) / cpi_base_year) * 100
    return inflation_rate
# Example price data for the base year and current year
base_year_data = [120, 150, 80, 60, 100]
current_year_data = [130, 160, 85, 65, 110]
inflation_rate = calculate_inflation(base_year_data, current_year_data)
print(f"Inflation Rate: {inflation_rate:.2f}%")
```
